```python
X = df_balanced.drop(columns='target_class')
y = df_balanced['target_class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train.isnull().sum().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='median')
X_train_arr = imputer.fit_transform(X_train)
X_test_arr  = imputer.transform(X_test)
```

```python
X_train_arr = pd.DataFrame(X_train_arr,columns=X_train.columns,index=X_train.index)

X_test_arr= pd.DataFrame(X_test_arr,columns=X_train.columns,index=X_test.index)
```

```python
type(X_test_arr)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_arr)
X_test_scaled = scaler.transform(X_test_arr)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns,index=X_train.index)

X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_train.columns,index=X_test.index)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(max_iter=1000)

log_reg.fit(X_train_scaled, y_train)
```

```python
y_pred = log_reg.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
